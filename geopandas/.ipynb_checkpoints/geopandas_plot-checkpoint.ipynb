```python
%matplotlib inline
import pandas as pd
import geopandas as gpd
```

```python
world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))#example gpd datasets
```

```python
world
```

```python
world.plot()
```

```python
af=world.loc[world["continent"] == "Africa"]
```

```python
af.plot(column="pop_est")
```

```python
af.plot(column="pop_est", cmap="terrain", figsize=(10,10))
# https://matplotlib.org/stable/tutorials/colors/colormaps.html
```

```python
na=world.loc[world["continent"] == "North America"]
na.plot()
```
