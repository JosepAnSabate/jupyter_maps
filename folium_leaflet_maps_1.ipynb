```python
import folium

m = folium.Map(location = [41.39247, 2.159637], zoom_start=15)
m
```

# Map styles

```python
j = folium.Map(location = [41.39247, 2.159637], tiles="Stamen Terrain", zoom_start=8)
j
```

# Marker

```python
folium.Marker(location=[41.39, 2.15], popup="<strong>Barcelona<strong/>", tooltip="Clica per a més info...").add_to(j)
j
```

```python
folium.Marker(location=[41.129849, 1.224701], tooltips="clica...", popup="Tarragona", icon=folium.Icon(icon='cloud')).add_to(j)
j
```

```python
folium.Marker(location=[41.371862, 1.182404], tooltips="clica...", popup="<Strong>Montblanc<Strong/>", icon=folium.Icon(icon='envelope', color="red")).add_to(j)
j
```

# Costumized markers

```python
locationFCB = [41.380959, 2.122983]
iconFCB = folium.features.CustomIcon('./images/fcb.png', icon_size=(60,60))
popupFCB = '<strong>Camp Nou</strong><br>Estadi del Futbol Club Barcelona'
folium.Marker(location=locationFCB, tooltips="Camp Nou", popup=popupFCB, icon=iconFCB).add_to(j)
j
```
