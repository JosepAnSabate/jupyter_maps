### Challenge 2: Open and Plot a CSV File with Time Series Data
The code above creates a path (stream_discharge_path) to open daily stream discharge measurements taken by U.S. Geological Survey from 1986 to 2013 at Boulder Creek in Boulder, Colorado. Using pandas, do the following with the data:

- 4. Plot the newly opened data with matplotlib. Make sure your x-axis is the dates, and your y-axis is the disValue column from the pandas DataFrame.

If you need a refresher on how to plot time series data, check out this lesson on working with time series data in Pandas.

```python
# Import data using datetime and no data value
precip_1986_2013_daily = pd.read_csv('C:/capes_sql/colorado-flood/discharge/06730200-discharge-daily-1986-2013.csv',
                                     parse_dates=['datetime'],
                                     index_col= ['datetime'],
                                     na_values=['999.99'],
                                    sep=',')

precip_1986_2013_daily.head()
```

```python
# Create figure and plot space
fig, ax = plt.subplots(figsize=(10, 10))

# Add x-axis and y-axis
ax.plot(precip_1986_2013_daily.index.values, #lineal = plot
           precip_1986_2013_daily['disValue'],
           color='purple')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Discharge (Cubic Feet per Second)",
       title="Stream Deischarge for Boulder Creek\n 1986-2013")

# Define the date format
date_form = DateFormatter("%Y")
ax.xaxis.set_major_formatter(date_form)

plt.show()
```

## Challenge 3: Subset the Data
The 2013 Colorado Flood occurred in 2013 (as the name implies). The plot above shows all of the stream discharge data over several decades. In this challenge you will subset the data to just the year and months during which the flood event occurred.

Do the following:

- 1.Subset the data to include only discharge data from August 1st, 2013 through October 31, 2013
- 2.Plot the newly subset data with matplotlib. Make sure your x-axis contains dates, and your y-axis is contains the disValue column from your pandas DataFrame.
- 3.Give your plot a title and label the axes.
- 4.Format the dates on the x-axis so they only show the month and the day. Additionally, you can angle the dates using the line of code fig.autofmt_xdate().
- 5.Make the x-axis week ticks only show up for every other week.

The lessons below should help you complete this challenge:

```python
#1Subset the data to include only discharge data from August 1st, 2013 through October 31, 2013
precip_2013 = precip_1986_2013_daily['2013-08-01':'2013-10-31']
```

```python
#2.Plot the newly subset data with matplotlib. Make sure your x-axis contains dates, and your y-axis is contains the disValue column from your pandas DataFrame.
#3.Give your plot a title and label the axes.
#4.Format the dates on the x-axis so they only show the month and the day. Additionally, you can angle the dates using the line of code
#5.Make the x-axis week ticks only show up for every other week.

# Create figure and plot space
fig, ax = plt.subplots(figsize=(10, 10))

# Add x-axis and y-axis
ax.scatter(precip_2013.index.values, #lineal = plot
           precip_2013['disValue'],
           color='purple')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Discharge (Cubic Feet per Second)",
       title="Stream Deischarge for Boulder Creek\n August - October 2013")

# Define the date format
date_form = DateFormatter("%b-%d")
ax.xaxis.set_major_formatter(date_form)

# 5
# For tickmarks and ticklabels every other week
# Ensure a major tick for each week using (interval=1) 
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))


plt.xticks(rotation = 45) # Rotates X-Axis Ticks by 45-degrees
plt.show()
```

## Challenge 4: Resample the Data
Next, summarize the stream discharge data by week. Additionally, you will clean up the format of the date labels on the x-axis. Do the following:

- 1.Resample the DataFrame that you made above for August - October, 2013 to represent the maximum stream discharge value for each week.
- 2.Plot the newly resampled data as a scatterplot (ax.scatter())using matplotlib. Give your plot a title and label the x and y axes.
- 3. Format the dates on the x-axis so they only show the month and the day. Additionally, you can angle the dates using the line of code fig.autofmt_xdate().
-4. Adjust the x-axis ticks and labels so you have one label and tick for every other week.


```python
# 1.Resample the DataFrame that you made above for August - October, 2013 to represent the maximum stream discharge value for each week
ppw=precip_2013.resample('W-MON').sum()
ppw.head()
```

```python
#2.Plot the newly resampled data as a scatterplot (ax.scatter())using matplotlib. Give your plot a title and label the x and y axes.
#3.Format the dates on the x-axis so they only show the month and the day. Additionally, you can angle the dates using the line of code fig.autofmt_xdate().
#4. Adjust the x-axis ticks and labels so you have one label and tick for every other week.

# Create figure and plot space
fig, ax = plt.subplots(figsize=(10, 10))

# Add x-axis and y-axis
ax.scatter(ppw.index.values, #lineal = plot
          ppw['disValue'],
           color='purple')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Discharge (Cubic Feet per Second)",
       title="Stream Deischarge for Boulder Creek\n August - October 2013")

# Define the date format
date_form = DateFormatter("%b-%d")
ax.xaxis.set_major_formatter(date_form)

plt.xticks(rotation = 45) # Rotates X-Axis Ticks by 45-degrees

# Ensure a major tick for each week using (interval=1) 
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
```

```python
# 1
precip_2003 = precip_1986_2013_daily['2003-08-01':'2003-10-31']
precip_2013 = precip_1986_2013_daily['2013-08-01':'2013-10-31']
```

```python
#2.Plot the data from 2003 on a plot above of the data from 2013 using matplotlib.
#3.Add titles to each plot and and label the x and y axes.
#6.Format the dates on the x-axis
f, (ax1, ax2) = plt.subplots(2,1, figsize=(12,9))# ,sharex=True You can use sharex or sharey to align the horizontal or vertical axis.

# Add x-axis and y-axis
ax1.scatter( precip_2003.index.values,
        precip_2003['disValue'],
       color='green')


# Set title and labels for axes
ax1.set(
       ylabel="Discharge (Cubic Feet per Second)",
       title="Stream Deischarge for Boulder Creek\n August - October 2003")

ax1.set(ylim=[0, 5000])#5
ax2.scatter(precip_2013.index.values,
       precip_2013['disValue'],
       color='purple')

# Set title and labels for axes
ax2.set(xlabel="Date",
       ylabel="Discharge (Cubic Feet per Second)",
       title="Stream Deischarge for Boulder Creek\n August - October 2013")

# Define the date format
date_form = DateFormatter("%b-%d")
ax1.xaxis.set_major_formatter(date_form)
ax2.xaxis.set_major_formatter(date_form)



# Ensure a major tick for each week using (interval=1) 
ax1.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
ax2.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))

ax1.axes.xaxis.set_ticklabels([])#drop x axis en sublots o plots

plt.tight_layout()#espaiat entre subplots
```
